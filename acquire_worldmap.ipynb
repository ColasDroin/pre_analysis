```python
import geopandas as gpd
import country_converter as coco
```

```python
# Load a low-resolution world shapefile
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

# Set France to iso_a3 'FRA'
world.loc[world['name'] == 'France', 'iso_a3'] = 'FRA'

# Set Norway to iso_a3 'NOR'
world.loc[world['name'] == 'Norway', 'iso_a3'] = 'NOR'

# Set Kuwait to iso_a3 'KWT'
world.loc[world['name'] == 'Kuwait', 'iso_a3'] = 'KWT'

# Set N. Cyprus to iso_a3 'CYP'
world.loc[world['name'] == 'N. Cyprus', 'iso_a3'] = 'CYP'

# Set Somaliland to iso_a3 'SOM'
world.loc[world['name'] == 'Somaliland', 'iso_a3'] = 'SOM'

# Set Kosovo to iso_a3 'RKS'
world.loc[world['name'] == 'Kosovo', 'iso_a3'] = 'XKX'

# Get iso_a2 from iso_a3
world['iso_a2'] = coco.convert(names=world['iso_a3'], to='ISO2')

# Remove Antarctica
world = world[(world['name'] != 'Antarctica')]
world
```

```python
# Select relevant columns: ISO_A2 and geometry
countries = world[["iso_a2", "geometry", "name"]]

# Filter out entries without ISO_A2 codes (if any)
countries = countries[countries["iso_a2"] != "-99"]

# Rename name to full_name
countries = countries.rename(columns={"name": "full_name"})

# Rename "iso_a2" to "name"
countries = countries.rename(columns={"iso_a2": "name"})

# Save to GeoJSON
output_path = "output/world_map.json"
countries.to_file(output_path, driver="GeoJSON")
```
